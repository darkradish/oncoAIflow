## **Class to translate from PDF documents to Dictionary**

```python
class Patient:
    def __init__(self, cirrhose, gastroscopie, ascite, encephalopathie, alcool, histoire_de_la_maladie, imc):
        self.cirrhose = cirrhose
        self.gastroscopie = gastroscopie
        self.ascite = ascite
        self.encephalopathie = encephalopathie
        self.alcool = alcool
        self.histoire_de_la_maladie = histoire_de_la_maladie
        self.imc = imc

    def to_dict(self):
        return {
            "cirrhose_documentee": self.cirrhose,
            "gastroscopie": {
                "oui_date_de_la_derniere": self.gastroscopie['oui_date_de_la_derniere'],
                "non": self.gastroscopie['non'],
                "Varices_oesophagiennes_gastriques": {
                    "Non_recherchees": self.gastroscopie['Varices_oesophagiennes_gastriques']['Non_recherchees'],
                    "Absentes": self.gastroscopie['Varices_oesophagiennes_gastriques']['Absentes'],
                    "Presentes_grade_des_VO": self.gastroscopie['Varices_oesophagiennes_gastriques']['Presentes_grade_des_VO']
                }
            },
            "ascite": {
                "Absente": self.ascite['Absente'],
                "Moderee": self.ascite['Moderee'],
                "Abondante": self.ascite['Abondante'],
                "Pas_ascite_clinique_mais_visible_en_imagerie": self.ascite['Pas_ascite_clinique_mais_visible_en_imagerie'],
                "Pas_ascite_actuellement_mais_ATCD_d_ascite": self.ascite['Pas_ascite_actuellement_mais_ATCD_d_ascite']
            },
            "encephalopathie": {
                "oui": self.encephalopathie['oui'],
                "non": self.encephalopathie['non'],
                "Si_oui_date_du_diagnostic": self.encephalopathie['Si_oui_date_du_diagnostic']
            },
            "alcool": {
                "OUI_depuis_combien_de_temps": self.alcool['OUI_depuis_combien_de_temps'],
                "NON_consommation_actuelle_d_alcool": self.alcool['NON_consommation_actuelle_d_alcool']
            },
            "histoire_de_la_maladie": {
                "date_du_diagnostic": self.histoire_de_la_maladie['date_du_diagnostic'],
                "Prouve_histologiquement": self.histoire_de_la_maladie['Prouve_histologiquement'],
                "traitements_locoregionaux": self.histoire_de_la_maladie['traitements_locoregionaux'],
                "traitements_systemiques": self.histoire_de_la_maladie['traitements_systemiques']
            },
            "imc": {
                "IMC": self.imc['IMC'],
                "ECOG": self.imc['ECOG']
            }
        }
```

## **Trial of PDF reader for Checkboxes**

```python
# Import PDF reader library
from PyPDF2 import PdfReader
import re

# Open the PDF file and read the first page
with open("test_0.pdf", "rb") as file:
    # Create a PDF file reader
    reader = PdfReader(file)
    # Get the number of pages
    num_pages = len(reader.pages)
    # Print the number of pages
    print(num_pages)
    # Get the first page
    page1 = reader.pages[0]
    # Extract the text from the first page
    text1 = page1.extract_text()

    # Use regular expression to extract 2 words after "❑"
    values = re.findall(r'❑ (\w+)', text1)

    # Print the extracted values
    for value in values:
        print(value)
```
